# Reading Data

When working with Unstructured Grid datasets, the grid and data variables are often stored as seperate netCDF files. This means that there are multiple seperate files that need to be read in at any given time.

For example, the NOAA Geoflow project consisits of 4 files (1 grid file and 3 data files)

```
geoflow-small
│   grid.nc
│   v1.nc
│   v2.nc
│   v3.nc
```




````python
import uxarray as ux
````

````python
base_path = "../../test/meshfiles/geoflow-small/"
grid_path = base_path + "grid.nc"
var_names = ['v1.nc', 'v2.nc', 'v3.nc']
data_paths = [base_path + name for name in var_names]
````

````python
grid = ux.open_dataset(grid_path, data_paths[0], data_paths[1], data_paths[2])
````
