```python
import uxarray as ux

import hvplot.pandas
import holoviews as hv
```

```python
base_path = "../../test/meshfiles/ugrid/outCSne30/"
grid_path = base_path + "outCSne30.ug"
data_path = base_path + "outCSne30_vortex.nc"
uxds = ux.open_dataset(grid_path, data_path)
uxds
```

```python
gdf = uxds['psi'].to_geodataframe()
```

```python
hv.extension("bokeh")
type(gdf.hvplot.polygons(rasterize=True))
```

```python
type(gdf.hvplot.polygons(datashade=True))
```

```python
import datashader as ds
import datashader.transfer_functions as tf

from holoviews.operation.datashader import rasterize
```

```python
cvs = ds.Canvas(plot_width=900, plot_height=600)
agg = cvs.polygons(gdf, geometry='geometry', agg=ds.mean('psi'))
filled = tf.shade(agg)
filled
```
